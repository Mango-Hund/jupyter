# 三联图

```python
# 打开并读取数据
rowdata = open(r'TSNE.csv', "r")
reader = csv.reader(rowdata)
headers = next(reader)

loc_x,loc_y,clu_x,clu_y,ser_x,ser_y = [],[],[],[],[],[]

for row in reader:
    loc_x.append(float(row[1])*20037508.342789/180)
    loc_y.append(math.log(math.tan((90+float(row[2]))*math.pi/360))/(math.pi/180)*20037508.34789/180)
    clu_x.append(float(row[3]))
    clu_y.append(float(row[4]))
    #ser_x.append(list(range(1,366)))
    ser_x.append(pd.date_range('2017-1-1', periods=365))
    ser_y.append(list(row[5:]))
```

```python
# 输出文件
#output_file('play.html')
```

```python
# 定义数据源
source_1 = ColumnDataSource(data=dict(x=loc_x,y=loc_y,xp=clu_x,yp=clu_y,xq=ser_x,yq=ser_y))
TOOLS_1='box_select,lasso_select,pan,wheel_zoom,reset'
```

## 画地图

```python
# 将wgs84坐标投影转换为墨卡托坐标

def wgs84toWebMercator(lon,lat):
    x =  lon*20037508.342789/180
    y =math.log(math.tan((90+lat)*math.pi/360))/(math.pi/180)
    y = y *20037508.34789/180
    return x,y
x1,y1 = wgs84toWebMercator(lon=120.79560279980944,lat=30.623112899720564)
x2,y2 = wgs84toWebMercator(lon=122.03020320013769,lat=31.925845100091397)
```

```python
#绘制地图
from bokeh.tile_providers import CARTODBPOSITRON, get_provider
tile_provider = get_provider(CARTODBPOSITRON)
# range bounds supplied in web mercator coordinates
fig_01 = figure(tools=TOOLS_1 ,x_range=(x1, x2), y_range=(y1,y2),plot_width=600,plot_height=600)
fig_01.axis.visible = False
fig_01.add_tile(tile_provider)
c1 = fig_01.circle('x','y', source=source_1)
c1.nonselection_glyph = Circle(fill_color='gray',fill_alpha=0.1, line_color=None)
c1.selection_glyph = Circle(fill_color='orange',line_color=None)
```

## 画拓扑图

```python
fig_02 = figure(tools=TOOLS_1,plot_width=600,plot_height=600,x_axis_label='Topological relationship', y_axis_label='Topological relationship')  #,title='Logical topology distribution of different users based on power consumption patterns'
fig_02.title.text_font_size = '10pt'    #标题字体
fig_02.xaxis.axis_label_text_font_size = "15pt"  #坐标轴字体
fig_02.yaxis.axis_label_text_font_size = "15pt"
c2 = fig_02.circle('xp','yp', source=source_1)
c2.nonselection_glyph = Circle(fill_color='gray',fill_alpha=0.4, line_color=None)
c2.selection_glyph = Circle(fill_color='orange',line_color=None)
```

## 画时序数据

```python
fig_03 = figure(tools=TOOLS_1,plot_width=600,plot_height=600,x_axis_type='datetime',y_range=[0, 1500], x_axis_label='Time series (365 days)', y_axis_label='Daily power consumption (kWh)') #,title='Electricity consumption data of different users'
fig_03.title.text_font_size = '10pt'    #标题字体
fig_03.xaxis.axis_label_text_font_size = "15pt"  #坐标轴字体
fig_03.yaxis.axis_label_text_font_size = "15pt"
c3 = fig_03.multi_line(xs='xq',ys='yq',source=source_1)
c3.nonselection_glyph = MultiLine(line_color='gray',line_alpha=0.2)
c3.selection_glyph = MultiLine(line_color='orange')
```

# 混合模型图

```python
statistik = pd.read_csv(r'Statistik.csv')
statistik.head(4)
```

```python
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, RangeTool
#将所需数据放入同一数据源
sum_ = statistik['sum'].values
day = statistik['day'].values
night = statistik['night'].values
dates = pd.to_datetime(statistik['date'])
source_3 = ColumnDataSource(data=dict(date=dates, sum_=sum_, day=day, night=night))
#画图
fig_19 = figure(plot_height=300, plot_width=800, tools="xpan", toolbar_location=None,
                x_axis_type="datetime", x_axis_location="above",background_fill_color="#efefef",x_range=(dates[100],dates[150]))
#总数
fig_19.line('date', 'sum_',source=source_3,color = 'red',line_width=3,legend_label="sum")
fig_19.circle('date', 'sum_',source=source_3, fill_color='white', size=3,legend_label="sum")
#白天用电量
fig_19.line('date', 'day',source=source_3,color = 'orange',line_width=3,legend_label="day")
fig_19.triangle('date', 'day',source=source_3,fill_color='white',size=3,legend_label="day")
#夜晚用电量
fig_19.line('date', 'night',source=source_3,color = 'blue',line_width=3,legend_label="night")
fig_19.hex('date', 'night',source=source_3,fill_color='white',size=3,legend_label="night")

range_tool = RangeTool(x_range=fig_19.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select = figure(title="Drag the middle and edges of the selection box to change the range above",
                plot_height=130, plot_width=800, y_range=fig_19.y_range,
                x_axis_type="datetime", y_axis_type=None,tools="", toolbar_location=None, background_fill_color="#efefef")
select.line(x='date',y='sum_',source=source_3,color = 'red',line_width=1)
select.line(x='date',y='day',source=source_3,color = 'orange',line_width=1)
select.line(x='date',y='night',source=source_3,color = 'blue',line_width=1)
select.ygrid.grid_line_color = None
select.add_tools(range_tool)
select.toolbar.active_multi = range_tool

show(column(fig_19, select))
```

```python
from bokeh.transform import linear_cmap
from bokeh.util.hex import hexbin
from sklearn import mixture
df=pd.read_csv('GMM.csv')
data=df.iloc[:,:].values
#特征工程
X,Y=[],[]
j = -1
for weight in data[:,9:10]:  #把用电量Z当成一种权重，Z越高，该点（X，Y）的权重越大，我们就重复更多的数量
    j+=1
    for i in range(1,math.floor(weight/100),1):   #用电量Z是多少(向下取整)，就复制多少个点（X，Y）
        X.append(data[j, 1:2])
        Y.append(data[j, 2:3])
X_train = np.hstack([X, Y])  #把“加权”后的X，Y拼起来，形成坐标

# 训练加权高斯混合模型
clf = mixture.GaussianMixture(n_components=5, covariance_type='full',random_state=0)
clf.fit(X_train)

# 准备画图，画网格
x = np.linspace(121.45, 121.95, num=50)  # linspace()指定的间隔内返回均匀间隔的数字。
y = np.linspace(30.82, 31.4, num=50)
X, Y = np.meshgrid(x, y)  # np.meshgrid()—生成网格点坐标矩阵
XX = np.array([X.ravel(), Y.ravel()]).T  # ravel()：多维数组转换为一维数组，如果没有必要，不会产生源数据的副本

Z = -clf.score_samples(XX)+4.5   #重新缩放输出Z
Z = -np.log(Z)#重新缩放输出Z
Z = list(Z)
bins = hexbin(x, y, 0.01)

fig_4 = figure(title="Manual hex bin for 50000 points", tools="wheel_zoom,pan,reset",
           match_aspect=True, background_fill_color='#440154')
fig_4.grid.visible = False
fig_4.hex_tile(q="q", r="r", size=0.1, line_color=None, source=bins,
           fill_color=linear_cmap('Z', 'Viridis256', 0, max(Z)))
show(fig_4)
```

```python
from math import pi

import pandas as pd

from bokeh.io import show
from bokeh.models import BasicTicker, ColorBar, LinearColorMapper, PrintfTickFormatter
from bokeh.plotting import figure
from bokeh.sampledata.unemployment1948 import data

data['Year'] = data['Year'].astype(str)
data = data.set_index('Year')
data.drop('Annual', axis=1, inplace=True)
data.columns.name = 'Month'

years = list(data.index)
months = list(data.columns)

# reshape to 1D array or rates with a month and year for each row.
df = pd.DataFrame(data.stack(), columns=['rate']).reset_index()

# this is the colormap from the original NYTimes plot
colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
mapper = LinearColorMapper(palette=colors, low=df.rate.min(), high=df.rate.max())

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"

p = figure(title="US Unemployment ({0} - {1})".format(years[0], years[-1]),
           x_range=years, y_range=list(reversed(months)),
           x_axis_location="above", plot_width=900, plot_height=400,
           tools=TOOLS, toolbar_location='below',
           tooltips=[('date', '@Month @Year'), ('rate', '@rate%')])

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "7px"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3

p.rect(x="Year", y="Month", width=1, height=1,
       source=df,
       fill_color={'field': 'rate', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="7px",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d%%"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

#show(p)      # show the plot
```
